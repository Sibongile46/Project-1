```python
import librosa
import numpy as np

def preprocess_audio(audio_path, target_sr=16000, max_len=5):
    # Load the audio file and resample it
    audio, sr = librosa.load(audio_path, sr=target_sr)
    
    # Calculate the maximum number of samples (max_len seconds)
    max_samples = target_sr * max_len

    # Pad or trim the audio to the maximum length
    if len(audio) < max_samples:
        # Pad with zeros if the audio is shorter than max_len
        audio = np.pad(audio, (0, max_samples - len(audio)), mode='constant')
    else:
        # Truncate the audio if it is longer than max_len
        audio = audio[:max_samples]
    
    return audio
```

```python
# Create full file paths
df['full_path'] = df['path'].apply(lambda x: os.path.join(audio_dir, x))

# Apply the preprocess_audio function to each full_path
df['processed_audio'] = df['full_path'].apply(preprocess_audio)
```

```python
import json

# Convert NumPy arrays to lists for JSON serialization
def convert_to_serializable(data):
    serializable_data = []
    for item in data:
        # Convert the ndarray (audio) to a list and keep the transcription
        serializable_data.append({
            'audio': item['audio'].tolist(),  # Convert ndarray to list
            'text': item['text']              # Keep the transcription as-is
        })
    return serializable_data

# Create your dataset
dataset = [{'audio': audio, 'text': transcription} for audio, transcription in zip(df['processed_audio'], df['sentence'])]

# Convert to serializable format
serializable_dataset = convert_to_serializable(dataset)

# Save as a JSON file
with open('processed_dataset.json', 'w') as f:
    json.dump(serializable_dataset, f)
```

```python
import evaluate
import numpy as np

wer_eva=evaluate.load("wer")


def compute_metrics(pred):
    pred_logits=pred.predictions
    pred_ids=np.argmax(pred_logits, axis=-1)
    
    pred.label_ids[pred.label_ids==-100]=processor.tokenizer.pad_token_id
    
    pred_str=processor.batch_decode(pred_ids)
    label_str=processor.batch_decode(pred.label_ids, group_tokens=False)
    
    wer=wer_eva.compute(predictions=pred_str, references=label_str)
    return {"wer": wer}
```

```python
import torchaudio

def speech_file_to_array_fn(path, target_sampling_rate=16000):
    # Load the audio file
    speech_array, original_sampling_rate = torchaudio.load(path)
    
    # Resample the audio if necessary
    if original_sampling_rate != target_sampling_rate:
        resampler = torchaudio.transforms.Resample(orig_freq=original_sampling_rate, new_freq=target_sampling_rate)
        speech_array = resampler(speech_array)
    
    # Convert to NumPy array
    return speech_array.squeeze().numpy(), target_sampling_rate
```

```python
def prepare_dataset(batch):
    # Load and resample the audio
    audio, sr = speech_file_to_array_fn(batch['full_path'], target_sampling_rate=16000)
    
    # Process the audio with padding
    batch["input_values"] = processor(audio, sampling_rate=sr, return_tensors="pt", padding=True).input_values[0]
    
    # Tokenize the text (labels) with padding
    with processor.as_target_processor():
        batch["labels"] = processor(batch["sentence"], padding=True).input_ids
    
    return batch
```

```python
# Apply the function to your dataset
processed_dataset = df.apply(prepare_dataset, axis=1)
```
